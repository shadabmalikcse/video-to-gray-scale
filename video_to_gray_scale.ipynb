```python
!pip install opencv-python

from google.colab import files
uploaded = files.upload()

import cv2
import os

video_file = next(iter(uploaded))

# Create a VideoCapture object
cap = cv2.VideoCapture(video_file)

# Get the video frame width, height, and frames per second (fps)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object to save the grayscale video
output_file = 'grayscale_video.mp4'
out = cv2.VideoWriter(output_file, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height), isColor=False)

# Read and convert each frame to grayscale
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Write the grayscale frame to the output video
    out.write(gray_frame)

# Release the VideoCapture and VideoWriter objects
cap.release()
out.release()

from google.colab import files

files.download(output_file)
```
